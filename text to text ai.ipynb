```python
import ipywidgets as widgets
from IPython.display import display, Markdown

# Input fields
name = widgets.Text(description="Name:")
email = widgets.Text(description="Email:")
phone = widgets.Text(description="Phone:")
degree = widgets.Text(description="Degree:")
college = widgets.Text(description="College:")
skills = widgets.Textarea(description="Skills", placeholder="e.g. Python, Java, Power BI...")
projects = widgets.Textarea(description="Projects", placeholder="List your projects...")
internships = widgets.Textarea(description="Internships", placeholder="Company, Role, Duration")
achievements = widgets.Textarea(description="Achievements", placeholder="Awards, honors, etc.")

# Display the form
display(name, email, phone, degree, college, skills, projects, internships, achievements)

# Function to generate detailed resume
def generate_detailed_resume(name, email, phone, degree, college, skills, projects, internships, achievements):
    resume = f"""\
# üìÑ Resume ‚Äì {name}

**Contact Information**
üìß Email: {email}
üìû Phone: {phone}

---

## üéì Education
{name} is currently pursuing {degree} from {college}, where they have gained strong foundational knowledge in their core domain and related fields.

---

## üíº Internships
{name} has undertaken the following internship(s):
{internships}

These roles helped them gain hands-on industry exposure, improve collaboration skills, and apply their academic learning in real-world projects.

---

## üõ†Ô∏è Technical Skills
{name} is proficient in the following skills:
{skills}

These skills enable them to develop intelligent systems, write efficient code, analyze data, and build complete applications with confidence.

---

## üß™ Academic & Personal Projects
Here are some of the key projects completed by {name}:
{projects}

These projects reflect their creativity, technical ability, and strong problem-solving mindset.

---

## üèÖ Achievements & Awards
{name} has accomplished the following during their academic journey:
{achievements}

These achievements showcase their dedication, hard work, and continuous improvement in both academics and extracurricular pursuits.

---
"""
    return resume

# Button and output
button = widgets.Button(description="Generate Detailed Resume")
output = widgets.Output()

def on_button_click(b):
    with output:
        output.clear_output()
        resume_text = generate_detailed_resume(
            name.value, email.value, phone.value, degree.value, college.value,
            skills.value, projects.value, internships.value, achievements.value
        )
        display(Markdown(resume_text))
        with open("DetailedResume.txt", "w") as f:
            f.write(resume_text)
        print("‚úÖ Detailed Resume saved as DetailedResume.txt")

button.on_click(on_button_click)
display(button, output)
```
